##Scrapping_Functions

```python
def get_party(soup):
    try:
        party=soup.find("div",{"class": "row_con"})
        if(party):
            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                next_td_tag = next_td_tag.findNext()
                if(next_td_tag.name == "a"):
                  return next_td_tag.text
                if(next_td_tag.name == "table"):
                    break
            return 'no name'
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_party(soup))
```

```python
def get_district(soup):
    district=[]
    try:
        party=soup.find("div",{"class": "row_con"})
        if(party):

            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            
            while(True):
               
                next_td_tag = next_td_tag.findNext()
                if(next_td_tag.name == "tr") :
                  district.append(next_td_tag.text)
                if(next_td_tag.name == "table"):
                    break
            x=district[-1].split('\n')[2].strip()
            return x
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_district(soup))
```

```python
def get_name(soup):
    try:
        party=soup.find("div",{"class": "components-wrapper"})
        if(party):
            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                next_td_tag = next_td_tag.findNext()
                if(next_td_tag.name == "h2"):
                  return next_td_tag.text
                if(next_td_tag.name == "table"):
                    break
            return 'no name'
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_name(soup))
```

```python
def get_birthday(soup):
    details=[]
    try:
        party=soup.find("div",{"class": "top-mp-detail-2"})
        if(party):

            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                
                next_td_tag = next_td_tag.findNext()

                if(next_td_tag.name == "td"):
                  details.append(next_td_tag.text)
                if(next_td_tag.name == "h4"):
                    break
            x=details[-3].split('\n')[0].split(":")[1].strip()
            return x
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_birthday(soup))
```

```python
def get_marital_status(soup):
    details=[]
    try:
        party=soup.find("div",{"class": "top-mp-detail-2"})
        if(party):

            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                
                next_td_tag = next_td_tag.findNext()

                if(next_td_tag.name == "td"):
                  details.append(next_td_tag.text)
                if(next_td_tag.name == "h4"):
                    break
            x=details[-2].split('\n')[0].split(":")[1].strip()
            return x
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_marital_status(soup))
```

```python
def get_religion(soup):
    details=[]
    try:
        party=soup.find("div",{"class": "top-mp-detail-2"})
        if(party):

            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                
                next_td_tag = next_td_tag.findNext()

                if(next_td_tag.name == "td"):
                  details.append(next_td_tag.text)
                if(next_td_tag.name == "h4"):
                    break
            x=details[-1].split('\n')[0].split(":")[1].strip()
            return x
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_religion(soup))
```

```python
def get_telephone(soup):
    details=[]
    try:
        party=soup.find("td",{"style": "padding-top:0;"})
        if(party):
            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            while(True):
                
                next_td_tag = next_td_tag.findNext()

                if(next_td_tag.name == "td"):
                  details.append(next_td_tag.text)
                if(next_td_tag.name == "h4"):
                    break
            if len(details[1])==10 and details[1][0]=='0':
              return details[1]
            return "අංකයක් නැත"
        else:
            print('No details')
            return 'error'   
    except Exception as e:
        print('ERROR get details',e)
        return 'error'
# print(get_telephone(soup))
```

```python
def get_email(soup):
    details=[]
    try:
        party=soup.find("a",{"style": "cursor: pointer;"})
        if(party):
            h2_1 = party.parent
            next_td_tag = h2_1.findNext()
            # print(next_td_tag.text)
            details.append(next_td_tag.text)
            while(True):
                details.append(next_td_tag.text)
                next_td_tag = next_td_tag.findNext()

                if(next_td_tag.name == "td"):
                  details.append(next_td_tag.text)
                if(next_td_tag.name == "td"):
                    break
                if len(details)>=1:
                  return details[1]
            if len(details)>=1:
                return details[1]
            return "විද්‍යුත් තැපෑලක් නැත"
        else:
            # print('No details')
            return 'error'   
    except Exception as e:
        # print('ERROR get details',e)
        return 'error'
# print(get_email(soup))
```
